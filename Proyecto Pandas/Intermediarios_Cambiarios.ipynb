```python
import pandas as pd

#Se realiza el proceso de web scraping
df_web = pd.read_html('https://gee.bccr.fi.cr/IndicadoresEconomicos/Cuadros/frmConsultaTCVentanilla.aspx')
```

```python
df_web[2]
```

```python
#Importe 38 registros y 6 columnas, debo eliminar dos registros mal importados
print(df_web[2].shape)

#Guardando el dtaframe en una variable nueva
df = df_web[2].copy()
del(df_web) #Elimino el df importado
```

```python
#Imprime encabezado de df
df.head(3)
```

```python
#Imprime los titulos de la columna
df.columns
```

```python
#Extraer fila o
primera_fila = df.iloc[0]

#Asignar las etiquetas a las columnas
df.columns = primera_fila

#borrar fila
df.drop(0, inplace=True)

df.head()
```

```python
#Renombrar las columnas
df.rename(columns={'Tipo de Entidad':'Tipo','Entidad Autorizada':'Entidad','Diferencial Cambiario':'Diferencial'}, inplace=True)

df.columns
```

```python
#Eliminar series con algun valor nulo
df.dropna()
```

```python
#Completa valores nulos con un valor especifico
df['Tipo'].fillna('NO VALOR').tail()
```

```python
df['Tipo'].ffill(inplace=True)
```

```python
#Opcion1 eliminar con index
df.drop(df.tail(1).index)

#Opcion2 eliminar con dropna
df.dropna(inplace=True)
```

```python
#Conultar la entidad con venta mal alta
tipo_venta_max = df['Venta'].max()


entidades_con_max = df[df['Venta'] == tipo_venta_max]


entidades_con_max
```

```python
entidades_con_max = df[df['Venta'] == tipo_venta_max]['Entidad']
entidades_con_max
```

```python
#Tipado de columnas
columnas_float = ['Compra','Venta','Diferencial']
df[columnas_float] = df[columnas_float].apply(pd.to_numeric)
print(df.dtypes)
```
